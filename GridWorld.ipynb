```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from operator import sub
output_notebook()
```

```python
class Agent(object):
    def __init__(self):
        self.rules = {'w':(0, -1), 'd':(1, 0), 's':(0, 1), 'a':(-1, 0)}
        self.pos = (0, 0)
        
    def move(self, x):
        if self.pos == (1, 0):
            self.pos = (1, 4)
            return 10
        if self.pos == (3, 0):
            self.pos = (3, 2)
            return 5
        
        pos_ = self.pos
        self.pos = tuple(map(sub, self.pos, self.rules[x]))
        if self.pos[0] < 0 or self.pos[0] > 4 or self.pos[1] < 0 or self.pos[1] > 4:
            self.pos = pos_
            return -1
        return 0        
```

```python
actions = ['w', 's', 'a', 'd']
gamma = 0.9
agent = Agent()
```

```python
# Bellman Equation
v = np.zeros((5, 5))
for epoch in range(100):
    for i in range(5):
        for j in range(5):
            v_ = 0
            for act in actions:
                agent.pos = (i, j)
                r = agent.move(act)
                x_, y_ = agent.pos
                v_ += (r + gamma * v[x_, y_])
            v[i, j] = v_ / 4
print v.transpose()
```

```python
p = figure(x_range=[0, 5], y_range=[0, 5])
p.image(image=[np.rot90(v)], x=[0], y=[0], dw=[5], dh=[5], palette="Spectral11")
show(p)
```

```python
# Optimal Value Functions
v = np.zeros((5, 5))
for epoch in range(100):
    for i in range(5):
        for j in range(5):
            v_ = []
            for act in actions:
                agent.pos = (i, j)
                r = agent.move(act)
                x_, y_ = agent.pos
                v_.append((r + gamma * v[x_, y_]))
            v[i, j] = max(v_)
print v.transpose()
```

```python
p = figure(x_range=[0, 5], y_range=[0, 5])
p.image(image=[np.rot90(v)], x=[0], y=[0], dw=[5], dh=[5], palette="Spectral11")
show(p)
```
